```python
import matplotlib.pyplot as plt
import numpy as np

# Define the time of day (24 hours, from 5 AM to 5 AM)
time_of_day = [f"{(i + 5) % 24}:00" for i in range(24)]

# Original data
Pload_arr = np.array([169, 175, 179, 171, 181, 190, 270, 264, 273, 281, 300, 320, 280, 260, 250, 200, 180, 190, 240, 280, 325, 350, 300, 250])
Hload_arr = np.array([130, 125, 120, 120, 125, 135, 150, 160, 175, 190, 195, 200, 195, 195, 180, 170, 185, 190, 195, 200, 195, 190, 180, 175])
Cload_arr = np.array([100, 100, 80, 100, 120, 135, 150, 135, 125, 130, 140, 150, 150, 130, 120, 110, 90, 80, 135, 150, 135, 140, 110, 125])

# Scaling factors
scaling_factors = [1, 0.5, 2]

# Create and plot three different scenarios for Electric, Heat, and Cooling loads
for i, scaling_factor in enumerate(scaling_factors):
    Pload_scenario = Pload_arr * scaling_factor
    Hload_scenario = Hload_arr * scaling_factor
    Cload_scenario = Cload_arr * 1

    # Create a new figure and plot the data
    plt.figure(figsize=(12, 6))

    plt.plot(time_of_day, Pload_scenario, marker='o', label=f'Electric Load (MG {i + 1})')
    plt.plot(time_of_day, Hload_scenario, marker='s', label=f'Heat Load (MG {i + 1})')
    plt.plot(time_of_day, Cload_scenario, marker='^', label=f'Cooling Load (MG {i + 1})')

    plt.xlabel('Time of Day',fontsize=20)
    plt.ylabel('Load',fontsize=20)
    plt.title(f'Microgrid {i + 1} - Scaled Load Profiles',fontsize=20)
    plt.xticks(rotation=45,fontsize=12)
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.savefig("./MultiMG/"+str(i)+"mg.png")
    plt.show()
```
